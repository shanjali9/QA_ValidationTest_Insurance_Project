```python
import pandas as pd

df = pd.read_csv("insure.csv")

df.head()
```

```python
def test_schema():

    expected_columns = [
        "age","sex","bmi",
        "children","smoker",
        "region","charges"
    ]

    assert list(df.columns) == expected_columns

def test_no_null_values():

    assert df.isnull().sum().sum() == 0

def test_no_duplicates():

    assert df.duplicated().sum() == 0

def test_age_range():

    assert df["age"].between(0,120).all()

def test_bmi_range():

    assert df["bmi"].between(10,60).all()

def test_sex_values():

    valid_values = ["male","female"]

    assert df["sex"].isin(valid_values).all()

def test_smoker_values():

    valid_values = ["yes","no"]

    assert df["smoker"].isin(valid_values).all()

def test_region_values():

    valid_regions = [
        "southwest",
        "southeast",
        "northwest",
        "northeast"
    ]

    assert df["region"].isin(valid_regions).all()

def test_charges_positive():

    assert (df["charges"] > 0).all()

def test_smoker_higher_charges():

    smokers = df[df["smoker"] == "yes"]["charges"].mean()
    nonsmokers = df[df["smoker"] == "no"]["charges"].mean()

    assert smokers > nonsmokers

df.duplicated().sum()
df[df.duplicated()]
df = df.drop_duplicates()
```

```python
test_schema()
test_no_null_values()
test_no_duplicates()
test_age_range()
test_bmi_range()
test_sex_values()
test_smoker_values()
test_region_values()
test_charges_positive()
test_smoker_higher_charges()

print("All tests passed")

print("Rows:", len(df))
print("Columns:", len(df.columns))
print("Duplicate rows:", df.duplicated().sum())
print("Missing values:", df.isnull().sum().sum())
print("Average charges:", df["charges"].mean())
```
